ランダムな数字の文

```python
#ランダムな整数を利用
import random
for i in range(1, 11):
  #変数numに１−２０までの整数を入れる
  num = random.randint(1,20)
  print(f"{i}回目の結果は{num}です。")
  #numの値が20なら強制終了する
  if num ==20:
    print("20が出たので繰り返し処理を強制終了します。")
    break
```

Continue

```python
for i in range(1, 11):
  # ループ変数iの値が奇数（2で割った余りが1）の時に、値を出力せずcontinue文で次のループに進む
  if i % 2 == 1:
    continue

    print (i)
```

```python
#ランダムな整数を利用するため、記述が必要です。
import random

# 変数numに0~4までのランダムな整数を代入する
num = random.randint(0, 4)

# 変数numの値が0以外である間、変数numの値を出力し続ける
while num != 0:
  num = random.randint(0, 4)
  print(num)
```

```python
# 引数$priceに送料を加算し、その値を出力する関数を定義する
def calculate_total(price):
    # 与えられた引数priceに消費税１０％を加算し、変数totalに代入して、整数にする
    total = price * 1.1
    return int(total)

    # 変数totalの値を出力する
    print(f"{total}円")

# 関数を呼び出し、引数として購入金額を渡す
calculate_total(110)
```

kadai_013_submit

```python
def calculate_total(price, tax_percent):
    # 消費税の割合を計算（10% → 0.1）
    tax_rate = tax_percent / 100

    # 消費税を加えた合計金額を求める
    total = price * (1 + tax_rate)
    return int(total)

    # 結果を戻り値として返す
    return total

# 関数を呼び出し、結果を出力する（課題の指定：110 と 10）
result = calculate_total(110, 10)
print(result)
```
