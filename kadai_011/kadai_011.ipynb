Break

```python
for i in range(1, 11):
  print (i)
  #変数の値が５ならbreak文で強制終了する
  if i ==5:
    break
```

ランダムな数字の文

```python
#ランダムな整数を利用
import random
for i in range(1, 11):
  #変数numに１−２０までの整数を入れる
  num = random.randint(1,20)
  print(f"{i}回目の結果は{num}です。")
  #numの値が20なら強制終了する
  if num ==20:
    print("20が出たので繰り返し処理を強制終了します。")
    break
```

Continue

```python
for i in range(1, 11):
  # ループ変数iの値が奇数（2で割った余りが1）の時に、値を出力せずcontinue文で次のループに進む
  if i % 2 == 1:
    continue

    print (i)
```

```python
#ランダムな整数を利用するため、記述が必要です。
import random

# 変数numに0~4までのランダムな整数を代入する
num = random.randint(0, 4)

# 変数numの値が0以外である間、変数numの値を出力し続ける
while num != 0:
  num = random.randint(0, 4)
  print(num)
```

課題（For繰り返し処理）

```python
# For文ループで9行表示
array = ["水","金","地","火","木","土","天","海","冥"]
for planet in array:
    print(planet)
```

課題（While繰り返し処理）

```python
# whileループで9行表示
array = ["水","金","地","火","木","土","天","海","冥"]

i = 0
while i < len(array):
    print(array[i])
    i += 1
```
